```python
#step function graph
def step_function(x):
    return np.array(x>0,dtype=np.int64)
x=np.arange(-5.0,5.0,0.1)
y=step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show
```

```python
#implement sigmoid funcion
def sigmoid(x):
    return 1/(1+np.exp(-x)) #np.exp가 넘파이 배열을 반환함 

x=np.array([-1.0,1.0,2.0])
sigmoid(x)
```

```python
#numpy broadcast
t = np.array([1.0,2.0,3.0])
1.0 / t
```

```python
#sigmoid function graph
x=np.arange(-5.0, 5.0, 0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show
```

```python
#compare step_function & sigmoid at graph
x=np.arange(-5.0, 5.0, 0.1)
y1=step_function(x)
y2=sigmoid(x)
plt.plot(x,y1,linestyle="--")
plt.plot(x,y2)
plt.ylim(-0.1,1.1)
plt.show
```

```python
#implement ReLU function
def relu(x):
    return np.maximum(0,x)

#ReLU funtion graph
x=np.arange(-5.0,5.0,0.1)
y=relu(x)
plt.plot(x,y)
plt.ylim(-0.5,5.5)
plt.show()
```
